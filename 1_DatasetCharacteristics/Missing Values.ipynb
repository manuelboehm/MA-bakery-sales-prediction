## Missing Values

```python
# Überprüfung des kompletten Datensatzes auf Missing Values

import pandas as pd

# Datensatz einlesen
df_dataset_complete = pd.read_csv("/workspaces/MA-bakery-sales-prediction/_data_prepared/df_dataset_complete.csv")  


# Filtere den Datensatz für den Zeitraum zwischen 01.07.2013 und 31.07.2018
filtered_data = df_dataset_complete[
    (df_dataset_complete['Datum'] >= '2013-07-01') & 
    (df_dataset_complete['Datum'] <= '2018-07-31')
]

# Berechne die Anzahl der fehlenden Werte pro Spalte
missing_values_filtered = filtered_data.isnull().sum()

# Ausgabe der Ergebnisse
print("Anzahl der fehlenden Werte pro Spalte (01.07.2013 bis 31.07.2018):")
print(missing_values_filtered)
```

```python
# Zeilen mit fehlendem Wert für 'Umsatz' anzeigen
missing_umsatz = filtered_data[filtered_data['Umsatz'].isnull()]

# Zeilen mit fehlendem Wert für 'Warengruppe' anzeigen
missing_warengruppe = filtered_data[filtered_data['Warengruppe'].isnull()]

# Zeilen mit fehlendem Wert für 'Brot' anzeigen
missing_brot = filtered_data[filtered_data['Brot'].isnull()]

# Wiederhole dies für die anderen Spalten, bei denen Werte fehlen
missing_broetchen = filtered_data[filtered_data['Broetchen'].isnull()]
missing_croissant = filtered_data[filtered_data['Croissant'].isnull()]
missing_konditorei = filtered_data[filtered_data['Konditorei'].isnull()]
missing_kuchen = filtered_data[filtered_data['Kuchen'].isnull()]
missing_saisonbrot = filtered_data[filtered_data['Saisonbrot'].isnull()]
missing_national_holiday = filtered_data[filtered_data['national_holiday'].isnull()]

# Ausgabe der ersten paar Zeilen für 'Umsatz'
print("Zeilen mit fehlendem Umsatz:")
print(missing_umsatz.head())

```

```python
# Zeilen mit fehlendem Wert für 'Umsatz' anzeigen
missing_umsatz = filtered_data[filtered_data['Umsatz'].isnull()]

# Ausgabe aller Zeilen, in denen der Umsatz fehlt
print("Alle Zeilen mit fehlendem Umsatz:")
print(missing_umsatz)
```
