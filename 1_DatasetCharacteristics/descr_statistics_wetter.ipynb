# Descriptive statistics wetter and wind

```python
# read in df_merged_umsatz_kiwo_wetter.csv (saved in 0_DataPreparation), create decriptive statistics for columns Bewoelkung, Temperatur, Windgeschwindigkeit, Wettercode; then create appropriate graphs 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# read in data 
df = pd.read_csv('/workspaces/MA-bakery-sales-prediction/0_DataPreparation/df_merged_umsatz_kiwo_wetter.csv')

# descriptive statistics for Bewoelkung
print(df['Bewoelkung'].describe())
print(df['Bewoelkung'].value_counts())
print(df['Bewoelkung'].unique())

# descriptive statistics for Temperatur
print(df['Temperatur'].describe())
print(df['Temperatur'].value_counts())
print(df['Temperatur'].unique())

# descriptive statistics for Windgeschwindigkeit
print(df['Windgeschwindigkeit'].describe())
print(df['Windgeschwindigkeit'].value_counts())
print(df['Windgeschwindigkeit'].unique())

# descriptive statistics for Wettercode
print(df['Wettercode'].describe())
print(df['Wettercode'].value_counts())
print(df['Wettercode'].unique())



# create boxplot for Bewoelkung
sns.set(style='whitegrid')
sns.boxplot(x='Bewoelkung', y='Umsatz', data=df)
plt.show()

# create boxplot for Temperatur
sns.set(style='whitegrid')
sns.boxplot(x='Temperatur', y='Umsatz', data=df)
plt.show()

# create boxplot for Windgeschwindigkeit
sns.set(style='whitegrid')
sns.boxplot(x='Windgeschwindigkeit', y='Umsatz', data=df)
plt.show()

# create boxplot for Wettercode
sns.set(style='whitegrid')
sns.boxplot(x='Wettercode', y='Umsatz', data=df)
plt.show()

# create scatterplot for Bewoelkung
sns.set(style='whitegrid')
sns.scatterplot(x='Bewoelkung', y='Umsatz', data=df)
plt.show()

# create scatterplot for Temperatur
sns.set(style='whitegrid')
sns.scatterplot(x='Temperatur', y='Umsatz', data=df)
plt.show()



```
